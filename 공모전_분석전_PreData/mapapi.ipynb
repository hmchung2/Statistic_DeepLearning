```python
import pandas as pd


df = pd.read_excel('star_two.xlsx',encoding='utf-8')
```

```python
len(df)
```

```python
df["location"]= df['location'].apply(lambda e: e.split(' (')[0])
```

```python
df["location"]
```

```python
locations = df["location"]
```

```python
locations[0]
```

```python
b = []
for i in locations:
    kor_url = quote(i)
    url = 'https://maps.googleapis.com/maps/api/geocode/json?address='+ kor_url +'&key=AIzaSyBu3tLXXVlXAcYi-hwPytWSjh6VakLsgHo&language=ko'
    req = Request(url, headers={ 'X-Mashape-Key': 'AIzaSyBu3tLXXVlXAcYi-hwPytWSjh6VakLsgHo' })
    ssltext = ssl.SSLContext(ssl.PROTOCOL_TLSv1)
    company_addr_json = urlopen(req, context=ssltext).read().decode('utf8')
    addr = json.loads(company_addr_json)
    b.append(addr['results'][0]['geometry']['location'])
```

```python
b
```

```python
b[0]
```

```python
locations[1]
```

```python
df = pd.DataFrame(list(zip(locations, b)), 
               columns =['starbucks_locations', 'x_y']) 
df 
```

```python
df2 = df['x_y']
```

```python
lat = []
lng = []
for i in range(len(df2)):
    lat.append(df2[i]['lat'])
    lng.append(df2[i]['lng'])
```

```python
lat
```

```python
lng
```

```python
starbucks_calc = pd.DataFrame(list(zip(lat, lng)), 
               columns =['lat', 'lng']) 
starbucks_calc
```
